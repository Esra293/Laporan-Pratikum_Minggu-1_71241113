```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
x = np.linspace(0, 10)
y = np.sin(x)
z = np.cos(x)
plt.plot(x, y, 'b', x, z, 'r')
plt.xlabel('radians')
plt.ylabel('value')
plt.title('Plotting Demonstration')
plt.legend(['Sin', 'Cos'])
plt.grid()
plt.show()
```

```python
def hitung_keuntungan(harga_beli_per_gram, harga_jual_per_gram, jumlah_emas):
    keuntungan_per_gram = harga_jual_per_gram - harga_beli_per_gram
    total_keuntungan = keuntungan_per_gram * jumlah_emas
    persentase_keuntungan = (total_keuntungan / (harga_beli_per_gram * jumlah_emas)) * 100
    return total_keuntungan, persentase_keuntungan

def keuntungan_awal():
    harga_beli_Awal = 650000
    harga_jual_awal = 685000
    jumlah_emas_awal = 25

    keuntungan_pertama, persentase_keuntungan_pertama = hitung_keuntungan(harga_beli_Awal, harga_jual_awal, jumlah_emas_awal)

    print(f"Keuntungan Awal: Rp. {keuntungan_pertama:,} ({persentase_keuntungan_pertama:.2f}%)")

def keuntungan_kedua():
    harga_beli_awal = 650000
    harga_beli_kedua = 685000
    harga_jual_kedua = 715000
    jumlah_emas_kedua = 40  
    harga_beli_total = (harga_beli_awal * 25 + harga_beli_kedua * 15) / 40
    keuntungan_kedua, persentase_keuntungan_kedua = hitung_keuntungan(harga_beli_total, harga_jual_kedua, jumlah_emas_kedua)

    print(f"Keuntungan kedua: Rp. {keuntungan_kedua:,} ({persentase_keuntungan_kedua:.2f}%)")
keuntungan_awal()
keuntungan_kedua()
```

```python
def waktu_tercapai(P, A, r):
    t = 0
    while P < A:
        P *= (1 + r) 
        t += 1  
    return t
P = 200000000  
A = 400000000  
r = 0.10 

waktu = waktu_tercapai(P, A, r)

print(f"Waktu yang dibutuhkan untuk mencapai 400 juta: {waktu} tahun")

```
